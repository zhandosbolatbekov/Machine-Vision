```python
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
im = pltimg.imread('6_26.jpg')
fig = plt.figure(figsize=(15,10))
plt.subplot(221)
plt.imshow(im[:,:,0])
plt.subplot(222)
plt.imshow(im[:,:,1])
plt.subplot(223)
plt.imshow(im[:,:,2])
plt.subplot(224)
plt.imshow(im)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
im = pltimg.imread('6_26.jpg')
fig = plt.figure(figsize=(15,10))
plt.subplot(221)
head = im[240:570,900:1200,:]
building = im[0:200, 0:250, :]
plt.imshow(im)
plt.subplot(222)
plt.imshow(head)
plt.subplot(223)
plt.imshow(building)
plt.show()
```
