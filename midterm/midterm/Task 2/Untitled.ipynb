```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.misc as img
from scipy import ndimage
import random

image = img.imread('lena_cover.jpg')
imag = img.imread('lena_cover.jpg', mode = 'F')

fig = plt.figure(figsize=(15,10))

plt.subplot(221)
plt.hist(image.ravel(), 512, [0,511])


plt.subplot(222)
plt.imshow(image)

rsum = 0 
gsum = 0
bsum = 0
for i in range(0, image.shape[0]):
    for j in range(0, image.shape[1]):
        rsum += image[i, j, 0]
        gsum += image[i, j, 1]
        bsum += image[i, j, 2]

rsum /= image.size
gsum /= image.size
bsum /= image.size
avg = (rsum + gsum + bsum) / 3
pr = avg / rsum
pg = avg / gsum
pb = avg / bsum

for i in range(0, image.shape[0]):
    for j in range(0, image.shape[1]):
        image[i, j, 0] *= pr
        image[i, j, 1] *= pg
        image[i, j, 2] *= pb

plt.subplot(223)
plt.imshow(image)

plt.show()
```
