```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.misc as img
from scipy import ndimage
import random

def sp_noise(image,prob):
    output = np.zeros(image.shape,np.uint8)
    thres = 1 - prob 
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            rdn = random.random()
            if rdn < prob:
                output[i][j] = 0
            elif rdn > thres:
                output[i][j] = 255
            else:
                output[i][j] = image[i][j]
    return output


image = img.imread('street.jpg' ,mode ='F')

spn_image = sp_noise(image, 0.03)

mdf_image = ndimage.median_filter(image2, 3)

fig = plt.figure(figsize=(15,10))

# original
plt.subplot(231)
plt.imshow(image,cmap='gray')

# salt-pepper noise
plt.subplot(232)
plt.imshow(spn_image, cmap='gray')

# median filtered 
plt.subplot(233)
plt.imshow(mdf_image, cmap='gray')

# difference between original and salt-pepper
plt.subplot(234)
adjusted = image - spn_image
plt.imshow(adjusted, cmap='gray')

# difference between adjusted and original
plt.subplot(235)
plt.imshow(adjusted - image, cmap='gray')

plt.show()
```
