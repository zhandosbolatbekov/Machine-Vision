```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.misc as img
from scipy import ndimage
from skimage import feature
import random

image = img.imread('lena_cover.jpg', mode='F' )

canny = feature.canny(image, sigma=20)

fig = plt.figure(figsize=(15,10))

plt.subplot(221)
plt.imshow(image, cmap='hot')

plt.subplot(222)
plt.imshow(canny, cmap='Reds')

sobel = ndimage.sobel(image)

plt.subplot(223)
plt.imshow(sobel, cmap='gray')

plt.show()
```
