```python
import matplotlib.pyplot as plt
import matplotlib.image as img
from scipy import ndimage as nimg 
import numpy as np
arms = img.imread('data2/Head/0.png')
gflt = nimg.median_filter(arms, 4)
fig = plt.figure(figsize=(15, 10))
plt.subplot(131)
plt.imshow(gflt, cmap='gray')
plt.subplot(132)
plt.imshow(arms, cmap = 'gray')
plt.imshow((arms - gflt), cmap = 'gray')
max_val = (arms - gflt).max()
min_val = (arms - gflt).min()
diff = np.log2(arms) - np.log2(gflt)
new_image = (diff - min_val) / (max_val - min_val)


plt.subplot(133)
plt.imshow(new_image, cmap='gray')
plt.show()
```
